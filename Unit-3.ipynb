# Getting image Resolution

   Resolution refers to the number of pixels (individual picture elements) contained in an image. It is typically expressed as the width by the height of the image in pixels. For example, an image with a resolution of 1920x1080 has a width of 1920 pixels and a height of 1080 pixels.

High Resolution: 
   Images with higher resolutions contain more pixels and therefore provide more detail. They are commonly used in print media or situations where a high level of detail is crucial.

Low Resolution: 
   Images with lower resolutions have fewer pixels, resulting in less detail. They are often used for web graphics, where smaller file sizes are desirable for faster loading times.

   The concept of resolution is essential in various fields, including photography, digital imaging, and computer graphics, as it directly impacts the clarity and quality of visual content.

# Converting Color Space

Color Space:

A color space is a specific organization of colors that allows for the representation and reproduction of a wide range of colors. It defines a specific mapping from an abstract color model to a concrete representation of colors in terms of intensity values or color components. Color spaces are used to describe how colors are represented in various mediums such as digital images, computer monitors, printers, and more.

In digital images, colors are typically represented using a combination of primary colors. The most common color model for digital images is the RGB (Red, Green, Blue) color model, where each pixel is represented by three values indicating the intensity of red, green, and blue. Other color spaces, such as HSV (Hue, Saturation, Value), LAB, and CMYK, are also used for specific purposes.

To perform color space conversion using scikit-image, you can use the color module, which provides various functions for working with color spaces. Here's an example of converting an RGB image to grayscale:

Converting Color Space:

Color space conversion is the process of transforming the representation of colors from one color space to another. This can be useful for various purposes, such as adjusting image properties, enhancing certain features, or preparing images for specific applications. Python libraries like OpenCV and scikit-image provide functions to perform color space conversions.

```python
from skimage import color
```

```python
grayimage = color.rgb2gray(img)
```

```python
io.imshow(grayimage)
```

```python
grayimage[:5,:5,]
```

The color space conversion from RGB to grayscale results in a single-channel image. In the RGB color space, each pixel is represented by three color channels (Red, Green, and Blue). When you convert an image to grayscale, you essentially combine these three channels into a single channel representing the intensity of light.

Therefore, after the conversion from RGB to grayscale using scikit-image's color.rgb2gray function, the resulting image will have only one channel. Each pixel in the grayscale image represents the luminance or brightness of the corresponding pixel in the original RGB image.

# Creating Basic Drawing 

The draw module in the scikit-image library provides various functions for drawing shapes and regions of interest on images

## Rectangle Function

                rectangle function:

                     Syntax: 

rr, cc = draw.rectangle(start, end, extent=None, shape=None)

              Draws a rectangle on an image.

                    Parameters:

 start: Tuple of starting coordinates (row, column).

 end: Tuple of ending coordinates (row, column).

 extent (optional): Extent of the rectangle (height, width).

 shape (optional): Shape of the image (rows, columns).

```python
width,height = 500,500
```

```python
import numpy as np
```

```python
image = np.zeros((width,height,3), dtype=np.uint8)
```

```python
from skimage import draw
```

```python
rr,cc = draw.rectangle(start = (0,400), end = (400,0))
```

```python
image[rr, cc] = [255, 0, 0]
```

```python
io.imshow(image)
```

            disk function:

Syntax: 
    
    rr, cc = draw.disk(center, radius, shape=None)
    
Draws a filled disk (circle) on an image.

center: Tuple of center coordinates (row, column).

radius: Radius of the disk.

## Doing Gamma Corrections 

Gamma correction is a technique used in image processing to adjust the brightness and contrast of an image. It involves applying a non-linear operation to the pixel values, specifically a power-law function, to compensate for the nonlinear way in which humans perceive light.

        I(corrected) = I(Original)*r

I(corrected) = Pixel intensity of corrected image

I(Original) = Pixel Intensity Of original image 

r = Gamma value

Here, γ is a parameter that controls the degree of correction. If γ is greater than 1, the image will be darkened. If γ is less than 1, the image will be brightened.

                                    Module: skimage.exposure

                                    Method: adjust_gamma

The adjust_gamma method just provides a convenient and flexible way to apply gamma correction in a single function call.

```python
img_path = r"C:\Users\C ARAVIND\Pictures\Saved Pictures\dark.jpg"
```

```python
dark = io.imread(img_path)
```

```python
from skimage import exposure 
```

```python
light = exposure.adjust_gamma(dark,0.7)
```

```python
io.imshow(light)
```

```python
io.imshow(dark)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(1,2,1)
plt.title("Original")
plt.axis("off")
plt.imshow(dark)

plt.subplot(1,2,2)
plt.title("Modified")
plt.axis("off")
plt.imshow(light)
```

#                                            Rotation 

                            Module: skimage.transform
                            
                                 Method: rotate

When you call skimage.transform.rotate(image, angle), the method takes in an input image (image) and the desired rotation angle (angle in degrees). It then applies a geometric transformation to the image, rotating each pixel by the specified angle around the image center.

The rotation is achieved using an affine transformation matrix. The matrix is calculated based on the rotation angle, and each pixel's coordinates are transformed according to this matrix. The interpolation algorithm is used to estimate the pixel values after rotation

```python
img_path = r"C:\Users\C ARAVIND\Pictures\Saved Pictures\normal.jpg"
```

```python
direction = io.imread(img_path)
```

```python
from skimage import transform
```

```python
io.imshow(direction)
```

```python
rotated = transform.rotate(direction,45)
```

```python
io.imshow(rotated)
```

```python
plt.subplot(1,2,1)
plt.title("Original")
plt.axis("off")
plt.imshow(direction)

plt.subplot(1,2,2)
plt.title("Modified")
plt.axis("off")
plt.imshow(rotated)

plt.show()
```

# Shifting 

                        Module: skimage.transform
                        
                            Class: AffineTransform
                            Method: translate
                            
he AffineTransform class in scikit-image allows you to perform affine transformations on images. The translate method within this class is specifically used for translation, which includes shifting an image along the x and y axes.

```python
from skimage.transform import AffineTransform
```

```python
shifted = transform.AffineTransform(translation = (200,-300))
```

```python
image_shifted = transform.warp(direction, shifted, mode='constant', cval=0)
```

```python
io.imshow(image_shifted)
```

```python
plt.subplot(1,2,1)
plt.title("Original")
plt.axis("off")
plt.imshow(direction)

plt.subplot(1,2,2)
plt.title("Modified")
plt.axis("off")
plt.imshow(image_shifted)

plt.show()
```

# Scalling

                            Module: skimage.transform
                            
                               Class: AffineTransform
                                Function: warp
                               
                                Uniform Scaling:

Uniform scaling involves resizing an image by the same factor along both the x and y axes. This can be achieved using the rescale function:

                              Non-uniform Scaling:

Non-uniform scaling allows different scaling factors for the x and y axes. This is achieved using the AffineTransform class:

                            skimage.transform.AffineTransform:
                            
The AffineTransform class represents an affine transformation, which includes operations like scaling, rotation, translation, and shearing.

The scale parameter in AffineTransform allows you to specify scaling factors for both the x and y axes.

                                skimage.transform.warp:
                                
The warp function applies a geometric transformation to an image.

It takes the input image and an affine transformation object as parameters.

The mode and cval parameters are optional and handle values outside the original image bounds.

```python
uniform_scaled = transform.rescale(direction, 1.3)
```

```python
io.imshow(uniform_scaled)
```

```python
non_linear = transform.AffineTransform(scale = (0.5,1))
```

```python
scaling = transform.warp(direction, non_linear)
```

```python
io.imshow(scaling)
```

# Structural Similarity

                            Module: skimage.metrics
                            
                         Method: structural_similarity

 1 indicates perfect similarity.

 0 indicates no similarity.
 
-1 indicates complete dissimilarity.

                                        Usage:

Call the structural_similarity function with the two images as arguments.

The function returns the structural similarity index.
