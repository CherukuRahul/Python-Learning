# Blending 

Blending two images using OpenCV involves combining the pixel values of the images to create a new image. You can achieve blending by assigning weights to each pixel in the images and combining them accordingly.

    cv2.addWeighted: This method is used for blending two images. It takes the following arguments:

        src1: The first input image.

        alpha: Weight of the first image elements.

        src2: The second input image.

        beta: Weight of the second image elements.

        gamma: Scalar added to each sum.

# Contrast and Brightness

                                                        Contrast:

Definition: Contrast refers to the difference in intensity between the darkest and lightest parts of an image. It is a measure of the spread of pixel values in an image.

High Contrast: In an image with high contrast, there is a significant difference between the dark and light areas. This 
results in a more visually striking and dynamic image.

Low Contrast: In an image with low contrast, the difference between dark and light areas is minimal. This can lead to a flatter and less visually interesting appearance.

                                                       Brightness:

Definition: Brightness refers to the overall lightness or darkness of an image. It represents the average intensity of all pixels in the image.

Increased Brightness: Increasing brightness makes the entire image lighter. This is achieved by adding a constant value to all pixel intensities.

Decreased Brightness: Decreasing brightness makes the entire image darker. This is achieved by subtracting a constant value from all pixel intensities.


                                            Contrast Adjustment Formula

                                      OUTPUT PIXEL = ALPHA * INPUT PIXEL + BETA 
                                      
                                            Brightness Adjustment Formula
                                            
                                          OUTPUT PIXEL = INPUT PIXEL + BETA
                                          
Adjusting these parameters allows you to control the visual impact of an image, making it suitable for different applications such as photography, computer vision, and image analysis.                                       

Here, alpha controls the contrast, and beta controls the brightness. You can achieve this using the cv2.convertScaleAbs function in OpenCV

# Adding Text To Image 

To add text to images using OpenCV in Python, you can use the cv2.putText function. 

text: The text string to be added.

font: The font type (e.g., cv2.FONT_HERSHEY_SIMPLEX).

font_scale: Font scale factor that multiplies the font-specific base size.

font_thickness: Thickness of the lines used to draw the text.

text_color: The color of the text in BGR format.

text_position: The position where the text should be placed, specified as (x, y) coordinates.

```python
from skimage import io
```

```python
import numpy as np
```

```python
blank = np.zeros((500,500,3), dtype = np.uint8)
```

```python
io.imshow(blank)
```

```python
text = 'Hello, OpenCV!'
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1.0
font_thickness = 2
text_color = (0, 255, 0)  # BGR format (green in this case)
text_position = (50, 50)
cv2.putText(blank, text, text_position, font, font_scale, text_color, font_thickness)
cv2.imshow("xyz",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Smoothing Filters 

## Gaussian Filter 

                                 Kernel: 
The Gaussian blur is based on a Gaussian function, which is a bell-shaped curve. The convolution operation is performed with a Gaussian kernel. The kernel is a matrix that determines how the values of neighboring pixels contribute to the blur of a specific pixel.

                              Smoothing Effect:
The convolution with a Gaussian kernel results in a weighted average of neighboring pixel values. Pixels closer to the center have a higher weight, and those farther away have lower weights. This creates a smoothing effect while preserving the overall structure of the image.

                            Standard Deviation:
The standard deviation of the Gaussian determines the spread of the bell curve. A higher standard deviation leads to a wider and smoother blur.

Gaussian blur is a widely used smoothing filter that applies a Gaussian function to the image. It's effective in reducing high-frequency noise and details.

                                  syntax :
        
     cv2.GaussianBlur(image,(kernel_size, kernel_size),standard_deviation)
     
image: The input image.

(kernel_size, kernel_size): The size of the kernel used for convolution. It should be an odd number.

0: The standard deviation of the Gaussian kernel. If set to 0, OpenCV automatically calculates it based on the kernel size.

## Median Blur 

                            Rank Filtering: 
Median blur is a type of rank filtering. Instead of using a mathematical function like in Gaussian blur, it sorts the pixel values in the neighborhood and replaces the pixel being processed with the median value.

                           Noise Reduction: 
Median blur is particularly effective in reducing "salt-and-pepper" noise in an image. This noise consists of random black and white pixels, and the median value is less affected by extreme values, making it robust against such noise.

                        Preservation of Edges: 
Unlike Gaussian blur, median blur does not create a weighted average of neighboring pixels. As a result, edges and details are better preserved.

Median blur is effective in removing salt-and-pepper noise by replacing each pixel's value with the median value in its neighborhood.

                            Syntax
             cv2.medianBlur(image, kernel_size)
 

image: The input image.

kernel_size: The size of the neighborhood used for the median calculation. It should be an odd number.

```python
medianblur = cv2.medianBlur(blur,9)
```

```python
plt.imshow(medianblur)
```

## Bilateral Filter 

                Spatial and Intensity Considerations: 
The bilateral filter considers both spatial closeness and intensity similarity when smoothing an image. It aims to smooth regions with similar intensity while preserving edges and boundaries.

                        Diameter Parameter (d): 
The diameter parameter (d) defines the size of the pixel neighborhood used during filtering. A larger diameter includes more distant pixels in the filtering process.

                      Color and Space Sigmas: 
Bilateral filtering introduces two sigma parameters (sigma_color and sigma_space).

                            sigma_color: 
Controls the color dissimilarity. A higher value allows a broader range of colors to be considered similar.

                            sigma_space: 
Controls the spatial dissimilarity. A higher value means that pixels farther away from the central pixel have more influence on the filtering.

                          Edge Preservation: 
The bilateral filter is effective at preserving edges because it avoids smoothing across areas with abrupt intensity changes.

The bilateral filter preserves edges while smoothing other regions. It takes into account both spatial and intensity differences.

                              Syntax
    blurred_image = cv2.bilateralFilter(image, d, sigma_color, sigma_space)

image: The input image.

d: Diameter of each pixel neighborhood.

sigma_color: Filter sigma in the color space. A higher value means more 
color variation is allowed.

sigma_space: Filter sigma in the coordinate space. A higher value means pixels farther away from the central pixel will influence the smoothing more.

```python
bFilter = cv2.bilateralFilter(blur,20,75,75)
```

```python
plt.imshow(bFilter)
```
